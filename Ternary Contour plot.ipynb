```python
import plotly.figure_factory as ff
import numpy as np

test_data = np.array([[0.700, 0.15, 0.15, 5.66, 39.54, 908.31], 
                        [0.642, 0.192, 0.166, 4.38, 35.36, 894.8], 
                        [0.552, 0.198, 0.250, 5.68, 41.09, 898.89], 
                        [0.601, 0.15, 0.249, 5.91, 36.61, 901.91], 
                        [0.601, 0.249, 0.150, 6.6, 37.36, 905.1], 
                        [0.583, 0.217, 0.200, 6.44, 34.41, 902.95], 
                        [0.500, 0.250, 0.250, 6.56, 37.54, 901.91], 
                        [0.500, 0.250, 0.250, 6.56, 37.54, 901.91], 
                        [0.601, 0.249, 0.150, 6.6, 37.36, 905.1], 
                        [0.643, 0.150, 0.207, 6.55, 35.26, 929.36], 
                        [0.601, 0.150, 0.249, 5.91, 36.61, 901.91], 
                        [0.539, 0.250, 0.211, 9.08, 35.32, 887.68], 
                        [0.700, 0.150, 0.150, 5.66, 39.54, 908.31], 
                        [0.612, 0.217, 0.171, 5.3, 34.43, 908.49], 
                        [0.613, 0.184, 0.203, 5.92, 35.95, 895.5], 
                        [0.552, 0.198, 0.250, 5.68, 41.09, 898.89]])



# barycentric coords: (a,b,c)
a = test_data[:, 0]
b = test_data[:, 1]
c = test_data[:, 2]

v = test_data[:, 3]

fig = ff.create_ternary_contour(
    np.array([a, b, c]), v,
    pole_labels=['A', 'B', 'C'],
    ncontours=5,
    colorscale='Reds',
    showscale=True,
    title='Ternary Contour Plot For FFA',
    coloring="lines",
    showmarkers=True
    
)

fig.update_layout(annotations=[
    dict(
        showarrow=False,
        text="%.2f" % value,
        x=x,
        y=y,
        font=dict(color='black')
    )
    for value, x, y in zip(np.ravel(test_data), np.ravel(a), np.ravel(v))
])

fig.update_layout(
    title='Ternary Contour Plot For FFA',
    ternary=dict(
        sum=100,
        aaxis=dict(title='WCO', ticks='inside', tickfont=dict(size=10)),
        baxis=dict(title='WPO', ticks='inside', tickfont=dict(size=10)),
        caxis=dict(title='WAF', ticks='inside', tickfont=dict(size=10)),
    ),
)

fig.show()

```
